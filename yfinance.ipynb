```python
import pandas as pd
import datetime
import numpy as np
from os import walk
for root,dirs,files in walk(r'C:/quant/USdaydate/Stock',topdown=False):
    print(files)#这里我们可以得到所有的文件名称，files是个list
num = len(files)#改文件夹下所有文件的总数量
alldata = pd.DataFrame() #建立一个空的dataframe
for i in range(num):
    newdata = pd.read_csv(r'C:/quant/USdaydate/Stock/%s'%files[i])#读取每个excel文件中的数据
    alldata = alldata.append(newdata)#将每个excel中的数据存储到之前建好的空的dataframe中
alldata = alldata.sort_values(by=['ts_code','Date'],axis=0,ascending=[True,True])
alldata.fillna(0,inplace=True) #去除null

alldata.head(10)
```

```python
alldata = alldata.sort_values(by=['ts_code','Date'],axis=0,ascending=[True,True])

alldata.reset_index(drop=True,inplace=True)
latestday=alldata.Date.max()
alldata.loc[alldata['Date']==latestday,'now_adj']=alldata['Adj Close'] # 最新复权因子
alldata.now_adj=alldata.now_adj.fillna(method='bfill')

#alldata.loc[alldata['ts_code']==alldata['ts_code'].shift(1),'now_adj']=alldata.now_adj.shift(1) # 最新复权因子

#alldata['Close'] = alldata['Close'] * alldata['Adj Close'] / alldata.now_adj  # 前复权 当日收盘价 × 当日复权因子 / 最新复权因子
alldata.shape
alldata
```

```python
Xf=alldata[alldata['Open'].notna()]
Xf=Xf[Xf['ts_code'].notna()]

close=Xf.Close
diff, dea, my_macd = talib.MACD(close,
                                             fastperiod=12,
                                             slowperiod=26,
#获取均线的数据，通过timeperiod参数来分别获取 5,10,20 日均线的数据。
                                             signalperiod=9)
SMA5=talib.SMA(close,timeperiod=5)
SMA10=talib.SMA(close,timeperiod=10)
SMA20=talib.SMA(close,timeperiod=20)
SMA60=talib.SMA(close,timeperiod=60)
SMA120=talib.SMA(close,timeperiod=120)

EMA5=talib.EMA(close,timeperiod=5)
EMA10=talib.EMA(close,timeperiod=10)
EMA20=talib.EMA(close,timeperiod=20)
EMA60=talib.EMA(close,timeperiod=60)
EMA120=talib.EMA(close,timeperiod=120)

Xf['diff']=diff
Xf['dea']=dea
Xf['MACD']=my_macd
Xf['SMA5']=SMA5
Xf['SMA10']=SMA10
Xf['SMA20']=SMA20
Xf['SMA60']=SMA60
Xf['SMA120']=SMA120
Xf['EMA5']=EMA5
Xf['EMA10']=EMA10
Xf['EMA60']=EMA60
Xf['EMA120']=EMA120

Xf['EMA20']=EMA20
Xf['SMACD-K']=(diff-dea)/dea
Xf['C/S']=(close/SMA20-1)*100
Xf['M/L']=(SMA60/SMA120-1)*100
Xf['S/M']=(SMA20/SMA60-1)*100
Xf['BreakSignal-20D']=close/EMA20-1
Xf['Amount']=close*Xf.Volume
Xf.reset_index(drop=True, inplace=True)

Xf['ATR'] = talib.ATR(Xf.High.values,Xf.Low.values,Xf.Close.shift(1).values, timeperiod=20)

Xf.loc[Xf['ts_code'] == Xf['ts_code'].shift(1),'Chg1D']= (Xf.Close/Xf.Close.shift(1)-1)*100
Xf.loc[Xf['ts_code'] == Xf['ts_code'].shift(4),'Chg5D']= (Xf.Close/Xf.Close.shift(4)-1)*100
Xf.loc[Xf['ts_code'] == Xf['ts_code'].shift(19),'Chg20D']= (Xf.Close/Xf.Close.shift(19)-1)*100

Xf.fillna(0,inplace=True) #去除null
Xf.loc[(Xf['Close']>Xf['EMA20']) & (Xf['Close']>Xf['SMA20']) & (Xf['SMA20']>Xf['SMA20'].shift(19)),'KL1']=1 #破线拐头
Xf.loc[(Xf['EMA20']>Xf['EMA60']) & (Xf['Close']>Xf['EMA20'].shift(19)),'K2']=1 #双均线交叉
Xf.loc[(Xf['EMA20']>Xf['SMA60']) & (Xf['Close']>Xf['EMA20'].shift(19)),'K3']=1 
Xf.loc[(Xf['SMA20']>Xf['EMA60']) & (Xf['Close']>Xf['SMA20'].shift(19)),'K4']=1 
Xf.loc[(Xf['SMA20']>Xf['SMA60']) & (Xf['Close']>Xf['SMA20'].shift(19)),'K5']=1 

Xf.loc[(Xf['Close']<Xf['EMA20']) & (Xf['Close']<Xf['SMA20']) & (Xf['SMA20']<Xf['SMA20'].shift(19)),'KL1']=-1 #空头方向破线拐头
Xf.loc[(Xf['EMA20']<Xf['EMA60']) & (Xf['Close']<Xf['EMA20'].shift(19)),'K2']=-1 #双均线交叉
Xf.loc[(Xf['EMA20']<Xf['SMA60']) & (Xf['Close']<Xf['EMA20'].shift(19)),'K3']=-1 
Xf.loc[(Xf['SMA20']<Xf['EMA60']) & (Xf['Close']<Xf['SMA20'].shift(19)),'K4']=-1 
Xf.loc[(Xf['SMA20']<Xf['SMA60']) & (Xf['Close']<Xf['SMA20'].shift(19)),'K5']=-1 
Xf.fillna(0,inplace=True) #去除null
Xf['trendscore-L']=(Xf.KL1+Xf.K2+Xf.K3+Xf.K4+Xf.K5)

Xf.loc[(Xf['Close']>Xf['EMA10']) & (Xf['Close']>Xf['SMA10']) & (Xf['SMA10']>Xf['SMA10'].shift(19)),'KS1']=1 #破线拐头
Xf.loc[(Xf['EMA10']>Xf['EMA20']) & (Xf['Close']>Xf['EMA10'].shift(19)),'KK2']=1 #双均线交叉
Xf.loc[(Xf['EMA10']>Xf['SMA20']) & (Xf['Close']>Xf['EMA10'].shift(19)),'KK3']=1 
Xf.loc[(Xf['SMA10']>Xf['EMA20']) & (Xf['Close']>Xf['SMA10'].shift(19)),'KK4']=1 
Xf.loc[(Xf['SMA10']>Xf['SMA20']) & (Xf['Close']>Xf['SMA10'].shift(19)),'KK5']=1 

Xf.loc[(Xf['Close']<Xf['EMA10']) & (Xf['Close']<Xf['SMA10']) & (Xf['SMA10']<Xf['SMA10'].shift(19)),'KS1']=-1 #空头方向破线拐头
Xf.loc[(Xf['EMA20']<Xf['EMA20']) & (Xf['Close']<Xf['EMA10'].shift(19)),'KK2']=-1 #双均线交叉
Xf.loc[(Xf['EMA20']<Xf['SMA20']) & (Xf['Close']<Xf['EMA10'].shift(19)),'KK3']=-1 
Xf.loc[(Xf['SMA20']<Xf['EMA20']) & (Xf['Close']<Xf['SMA10'].shift(19)),'KK4']=-1 
Xf.loc[(Xf['SMA20']<Xf['SMA20']) & (Xf['Close']<Xf['SMA10'].shift(19)),'KK5']=-1 
Xf.fillna(0,inplace=True) #去除null
Xf['trendscore-S']=(Xf.KS1+Xf.KK2+Xf.KK3+Xf.KK4+Xf.KK5)

Xf.head(10)
```

```python
latestday=max(Xf.Date)

valve=50000000 #以5000万美元为界限
Xf2=Xf[(Xf.Date==latestday) & (Xf.Amount > valve)]
Xf2=Xf2[['Date','ts_code','Close','Amount','Chg1D','Chg5D','Chg20D','C/S','S/M','M/L','ATR','KL1','KS1','trendscore-S','trendscore-L']]

Xf2
```

```python
#找长期，中期乖离率高，但短期下杀严重，趋势却走好的
DF=Xf2[(Xf2['C/S']<0) & (Xf2['S/M']>3) & (Xf2['M/L']> 3) & (Xf2['trendscore-L']>=3)]
DF
```
